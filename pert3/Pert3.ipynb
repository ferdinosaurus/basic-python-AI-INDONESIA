```python
def prima(n):
    if n<1:
        return False
    for x in range(2,n):
        if n%x ==0:
            return False
    
    else:
        return True

n=4
if prima(n):
    print(f"{n} adalah bilangan prima")
else :
    print(f"{n} adalah bukan bilangan prima")
    
```

```python
def daftar_prima(n):
    for n in range(100):
        if prima(n):
            print(f"{n} adalah bilangan prima")
        else :
            print(f"{n} adalah bukan bilangan prima")

daftar_prima(1)
```
